```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import numpy as np
import requests
from bs4 import BeautifulSoup

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, auc, precision_recall_curve
)
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
detector = SentimentIntensityAnalyzer()
```

```python
# Paths (adjust if using Drive)
dataset_path = "/content/drive/MyDrive/FakeNewsProject/updated_news_dataset.csv"
vectorizer_path = "/content/drive/MyDrive/FakeNewsProject/tfidf_vectorizer.pkl"

# ---- Web scraping functions (same as lightweight) ----
def scrape_bbc(limit=20):
    try:
        url = "https://www.bbc.com/news"
        resp = requests.get(url, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        items = soup.find_all("h3")
        texts = [it.get_text(strip=True) for it in items[:limit]]
        return pd.DataFrame({"text": texts, "label": 0})
    except Exception as e:
        print("BBC scrape error:", e)
        return pd.DataFrame(columns=["text", "label"])

def scrape_cnn(limit=20):
    try:
        url = "https://edition.cnn.com/world"
        resp = requests.get(url, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        items = soup.find_all("span", class_="container__headline")
        texts = [it.get_text(strip=True) for it in items[:limit]]
        return pd.DataFrame({"text": texts, "label": 0})
    except Exception as e:
        print("CNN scrape error:", e)
        return pd.DataFrame(columns=["text", "label"])

def scrape_reuters(limit=20):
    try:
        url = "https://www.reuters.com/world/"
        resp = requests.get(url, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        items = soup.find_all(["h2","h3"])
        texts = []
        for it in items:
            t = it.get_text(strip=True)
            if len(t) > 5:
                texts.append(t)
            if len(texts) >= limit:
                break
        return pd.DataFrame({"text": texts, "label": 0})
    except Exception as e:
        print("Reuters scrape error:", e)
        return pd.DataFrame(columns=["text", "label"])

def get_all_real_news(limit_each=10):
    a = scrape_bbc(limit=limit_each)
    b = scrape_cnn(limit=limit_each)
    c = scrape_reuters(limit=limit_each)
    return pd.concat([a,b,c], ignore_index=True)
```

```python

# ---- Load dataset & append scraped real news ----
df = pd.read_csv(dataset_path)

# Append a few scraped real news items to enrich dataset, then sample
scraped = get_all_real_news(limit_each=10)   # small addition
if not scraped.empty:
    df = pd.concat([df, scraped], ignore_index=True)

# Sample a smaller fraction for heavy models
df = df.sample(frac=0.3, random_state=42).reset_index(drop=True)

# Load vectorizer
vectorizer = joblib.load(vectorizer_path)

# Raw texts and labels
X_raw = df["text"]
y = df["label"]

# Split raw first, then vectorize
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=42, stratify=y
)
X_train = vectorizer.transform(X_train_raw)
X_test = vectorizer.transform(X_test_raw)
```

```python
def evaluate_model(name, model, X_train, X_test, y_train, y_test, X_test_raw=None):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    print(f"\n=== {name} ===")
    print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f}")

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Real", "Fake"], yticklabels=["Real", "Fake"])
    plt.title(f"{name} - Confusion Matrix")
    plt.show()

    if y_prob is not None:
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC={roc_auc:.2f})'); plt.plot([0,1],[0,1],'k--')
        plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title(f"ROC Curve - {name}"); plt.legend(); plt.show()
        precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_prob)
        plt.plot(recall_vals, precision_vals, label=name); plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"Precision-Recall Curve - {name}"); plt.legend(); plt.show()

    avg_sent = np.nan; avg_sent_pred_fake = np.nan; avg_sent_pred_real = np.nan
    if X_test_raw is not None:
        sent_compounds = np.array([detector.polarity_scores(str(t))['compound'] for t in X_test_raw])
        avg_sent = sent_compounds.mean()
        mask_pred_fake = (y_pred == 1); mask_pred_real = (y_pred == 0)
        if mask_pred_fake.any(): avg_sent_pred_fake = sent_compounds[mask_pred_fake].mean()
        if mask_pred_real.any(): avg_sent_pred_real = sent_compounds[mask_pred_real].mean()
        print(f"Average sentiment (compound) on TEST set: {avg_sent:.4f}")
        print(f"Average sentiment for samples PREDICTED as Fake: {avg_sent_pred_fake}")
        print(f"Average sentiment for samples PREDICTED as Real: {avg_sent_pred_real}")

    return {"Model": name, "Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1,
            "Avg_Sentiment": avg_sent, "Avg_Sent_Pred_Fake": avg_sent_pred_fake, "Avg_Sent_Pred_Real": avg_sent_pred_real}
```

```python
heavy_models = {
    "SVM": SVC(probability=True, kernel="linear"),
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=50, eval_metric="logloss", use_label_encoder=False)
}

heavy_results = []
for name, model in heavy_models.items():
    res = evaluate_model(name, model, X_train, X_test, y_train, y_test, X_test_raw=X_test_raw)
    heavy_results.append(res)

df_heavy_results = pd.DataFrame(heavy_results)
print("\nðŸ“Š Heavy Models Results:\n", df_heavy_results)
df_heavy_results.to_csv("/content/heavy_model_results.csv", index=False)
```
